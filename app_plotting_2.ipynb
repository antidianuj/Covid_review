```python
import geopandas as gpd
shapefile = 'ne_110m_admin_0_countries.shp'
#Read shapefile using Geopandas
gdf = gpd.read_file(shapefile)[['ADMIN', 'ADM0_A3', 'geometry']]
#Rename columns.
gdf.columns = ['country', 'country_code', 'geometry']
gdf.head()
```

```python
import pandas as pd
import collections



datafile = 'COVID tracing apps data_23Jan2021_Junaid Iqbal Khan.xlsx'
#Read csv file using pandas
df = pd.read_excel(datafile)
df.head()
locations=df['Location'].to_list()


counter=collections.Counter(locations)
freq_locations=list(counter.values())

uniq_locations=list(counter.keys())


# Calling DataFrame constructor after zipping
# both lists, with columns specified
df = pd.DataFrame(list(zip(uniq_locations, freq_locations)),
               columns =['Location', 'Frequency'])

df.head(20)
```

```python
merged = gdf.merge(df, left_on = 'country', right_on = 'Location')
merged.head(80)
```

```python


import pandas as pd
import collections
import math 

def split(word):
    return [char for char in word]

datafile = 'COVID tracing apps data_23Jan2021_Junaid Iqbal Khan.xlsx'
#Read csv file using pandas
df = pd.read_excel(datafile)
df.head()
locations=df['Location'].to_list()
tech=df['Tech'].to_list()

counter=collections.Counter(locations)
freq_locations=list(counter.values())

uniq_locations=list(counter.keys())

moder= []

def notisNaN(string):
    return string == string


moder=[]
for i in range(len(uniq_locations)):
  for j in range(len(locations)):
    theta=[]
    if locations[j]==uniq_locations[i]:
          if notisNaN(tech[j]):
            if "B" in tech[j] or "b" in tech[j]: 
              theta.append(1)

            if "L" in tech[j] or "n" in tech[j]: 
              theta.append(2)

            if "Q" in tech[j]: 
              theta.append(3)

          else:
            theta.append(0)


          moder.append(theta)




Bt=[]
Other=[]
for i in range(len(moder)):
  if 1 in moder[i]:
    Bt.append(1)
  else:
    Bt.append(0)

for i in range(len(moder)):
  if 2 in moder[i] or 3 in moder[i]:
    Other.append(1)
  else:
    Other.append(0)



df = pd.DataFrame(list(zip(locations, Bt, Other)),
               columns =['Location', 'Bluetooth','Other'])

df.head(20)


```

```python
merged = gdf.merge(df, left_on = 'country', right_on = 'Location')
merged.head(80)
```

```python
import geoplot
geoplot.choropleth(merged, hue='Bluetooth', edgecolor='grey', linewidth=1,
    cmap='Blues', legend=True, figsize=(20, 10))
```

```python
import geoplot
geoplot.choropleth(merged, hue='Other', edgecolor='grey', linewidth=1,
    cmap='OrRd', legend=True, figsize=(20, 10))
```
